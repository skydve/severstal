```python
def winsorize(x, l, h):  
    UPPERBOUND, LOWERBOUND = np.percentile(x, [l, h])
    return np.clip(x, UPPERBOUND, LOWERBOUND)

def scale(df, scaler):
    return scaler.fit_transform(df)
```

```python
def plot_fi(fi): 
    return fi.plot('cols', 'imp', 'barh', figsize=(12,7), legend=False)
```

```python
def test_xgb(df, target):
    X_train, X_test, y_train, y_test = train_test_split(df, target, test_size=0.1, random_state=7)
    m = LGBMClassifier(max_depth=6, random_state=7)    
    m.fit(X_train, y_train)
    preds = m.predict_proba(X_test)
    print('AUC_SCORE:', roc_auc_score(y_test, preds[:, 1]))
    fi = rf_feat_importance(m, df)
    plot_fi(fi[:30])
```

**knn**

т.к. knn не еффективен на высоких размерностях, возьмем самые информативные признаки

```python
features = ['34', '37', '10', '42']
df2 = df[[*features, 'target']].copy()
df2.fillna(df2.median(), inplace=True)
# df2= pd.DataFrame(scaler.fit_transform(df2))
train, valid = df2[:800000], df2[800000:]
```

```python
%%time
scores = []
for n in tqdm([5, 10, 15, 20, 25, 30, 35, 40]):
    m = KNeighborsClassifier(n_neighbors=n, n_jobs=-1)
    m.fit(train[features], train['target'])
    preds = m.predict_proba(valid[features])
    scores.append(roc_auc_score(valid['target'], preds[:, 1]))    
```

**simple stacking**

```python
models = (LGBMClassifier(max_depth=11, n_estimators=150, learning_rate=0.18), \
          CatBoostClassifier(logging_level='Silent', max_depth=11, n_estimators=150, learning_rate=0.18), \
          XGBClassifier(max_depth=11, n_estimators=25, learning_rate=0.18, n_jobs=-1))
```

```python
scores = []
preds_stack = []
for model in tqdm(models):
    m = model
    m.fit(X_train, y_train)
    preds = m.predict_proba(X_test)
    preds_stack.append(preds[:, 1])
    scores.append(roc_auc_score(y_test, preds[:, 1]))
print(scores)
final_preds = 0.4*preds_stack[0] + 0.4*preds_stack[1] + 0.2*preds_stack[2]
print(roc_auc_score(y_test, final_preds))
```
